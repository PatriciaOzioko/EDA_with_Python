# Querying the Data

```python
# Checking for the number of records without any medal mentioned
records_without_medal = Athens['Medal'].isnull().sum()
print(f"Number of Records without Medal: {records_without_medal}")
```

```python
# Counting the most young and most old people who got Gold Medal
# Filtering for Gold Medalists
gold_medalists = Athens[Athens['Medal'] == 'Gold']

# Finding the youngest and oldest individuals
youngest_gold_medalist = gold_medalists.loc[gold_medalists['Age'].idxmin()]
oldest_gold_medalist = gold_medalists.loc[gold_medalists['Age'].idxmax()]

# Counting the number of youngest and oldest Gold Medalists
count_youngest = gold_medalists[gold_medalists['Age'] == youngest_gold_medalist['Age']].shape[0]
count_oldest = gold_medalists[gold_medalists['Age'] == oldest_gold_medalist['Age']].shape[0]

print(f"Number of youngest Gold Medalists: {count_youngest}")
print(f"Youngest Gold Medalist:\n{youngest_gold_medalist}\n")

print(f"Number of oldest Gold Medalists: {count_oldest}")
print(f"Oldest Gold Medalist:\n{oldest_gold_medalist}\n")
```

```python
# Count/Find number of gold medals won by women of any specific country in a particular year
# Lets choose a country and year of interest
desired_country = 'USA'
desired_year = 2016

# Filtering the DataFrame based on my country and year of choice
filtered_data = Athens[(Athens['Team'] == desired_country) & (Athens['Year'] == desired_year)]

# Further filtering to include only gold medals won by women
gold_medals_women = filtered_data[(filtered_data['Medal'] == 'Gold') & (filtered_data['Sex'] == 'F')]

# Count the number of gold medals won by women
count_gold_medals_women = gold_medals_women.shape[0]

print(f"Number of gold medals won by women of {desired_country} in {desired_year}: {count_gold_medals_women}")
```

# Cleaning and Completing Data

```python
# Excluding all records where information about data is missing
Athens_with_medals = Athens.dropna(subset=['Medal'])

print(Athens_with_medals)
```

```python
# Filling missing age values with average age of other athletes.
average_age = Athens_with_medals['Age'].mean()

# Fill missing age values with the average age
Athens_with_medals.loc[:,'Age'] = Athens_with_medals['Age'].fillna(average_age)

# Display with filled age values
print(Athens_with_medals)
```

```python
# Fill missing height values for women and men with average height of women and men athletes respectively.
avg_height_women = Athens_with_medals[Athens_with_medals['Sex']== 'F']['Height'].mean()
avg_height_men = Athens_with_medals[Athens_with_medals['Sex']== 'M']['Height'].mean()

#Filling missing height values for women and men with their respective averages
Athens_with_medals.loc[Athens_with_medals['Sex'] == 'F', 'Height'] = Athens_with_medals.loc[Athens_with_medals['Sex'] == 'F', 'Height'].fillna(avg_height_women)
Athens_with_medals.loc[Athens_with_medals['Sex'] == 'M', 'Height'] = Athens_with_medals.loc[Athens_with_medals['Sex'] == 'M', 'Height'].fillna(avg_height_men)

# Display with filled height values
print(Athens_with_medals)
```

```python
# Fill missing weight values for women and men with average weight of women and men athletes participating in same sports
# Grouping by 'Sport' and 'Sex' to calculate the average weight for women and men in each sport
avg_weight_by_sport_sex = Athens_with_medals.groupby(['Sport', 'Sex'])['Weight'].mean().reset_index()

# Using a function to fill missing weights based on the average weight in the same sport and sex
def fill_missing_weight(row):
    if pd.isnull(row['Weight']):
        try:
            return avg_weight_by_sport_sex[
                (avg_weight_by_sport_sex['Sport'] == row['Sport']) & 
                (avg_weight_by_sport_sex['Sex'] == row['Sex'])
            ]['Weight'].values[0]
        except IndexError:
            return row['Weight']
    else:
        return row['Weight']

# Apply the function to fill missing weights
Athens_with_medals.loc[:, 'Weight'] = Athens_with_medals.apply(fill_missing_weight, axis=1)

# Display the result with filled weight values
print(Athens_with_medals)
```

# Data Visualization

```python
# Count Plot for Gold medals in gymnastic over age
import seaborn as sns
import matplotlib.pyplot as plt

# Filtering for Gold Medals in Gymnastics
gymnastics_gold = Athens_with_medals[(Athens_with_medals['Sport'] == 'Gymnastics') & (Athens_with_medals['Medal'] == 'Gold')]

# Create a count plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Age', data=gymnastics_gold, palette='viridis')
plt.title('Count of Gold Medals in Gymnastics Over Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right' )
plt.show()
```

```python
# Plot shows that the age with the highest Gold Medals in the Gymnastic Sport is 24. Does this mean that atheletes at this age do better in this sport?
#  while the observation about the age of 24 is interesting, it should be seen as a starting point for further exploration rather than a strict guideline.
# Success in sports is influenced by a multitude of factors, and a balanced and individualized approach to athlete development is key. 
# Countries should instead focus on comprehensive talent identification, development programs, and support systems that consider the unique characteristics and potential of each athlete, regardless of their age.
```

```python
# Countplot for Medals won by China over years
# Filtering for Medals Won by China
china_medals = Athens_with_medals[Athens_with_medals['Team'] == 'China']

# Create a count plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=china_medals, palette='Set1')
plt.title('Count of Medals Won by China Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# China has won a good number of medals with they lowest being in 1994 and the highest medals won was in 2008. What caused the decline after this year can be investigated.
```

```python
# Countplot for Gold medals won by China in summer Olympics in sports
# Filtering for Gold Medals Won by China in Summer Olympics
china_gold_summer = Athens_with_medals[(Athens_with_medals['Team'] == 'China') & 
                                     (Athens_with_medals['Season'] == 'Summer') & 
                                     (Athens_with_medals['Medal'] == 'Gold')]

# Create a count plot
plt.figure(figsize=(12, 6))
sns.countplot(x='Sport', data=china_gold_summer, palette='pastel', order=china_gold_summer['Sport'].value_counts().index)
plt.title('Count of Gold Medals Won by China in Summer Olympics in Sports')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Pointplot for Height of Male and Female Athletes over years
# Filtering for male and female athletes
male_athletes = Athens_with_medals[Athens_with_medals['Sex'] == 'M']
female_athletes = Athens_with_medals[Athens_with_medals['Sex'] == 'F']

# Creating point plots for height over years for male and female athletes
plt.figure(figsize=(14, 7))

# Point plot for male athletes
sns.pointplot(x='Year', y='Height', data=male_athletes, color='blue', label='Male Athletes', errorbar=None)

# Point plot for female athletes
sns.pointplot(x='Year', y='Height', data=female_athletes, color='pink', label='Female Athletes', errorbar=None)

# Add labels and title
plt.title('Height of Male and Female Athletes Over Years')
plt.xlabel('Year')
plt.ylabel('Height')
plt.legend()  

plt.show()
```

```python
#Barplot for Top 5 countries with most medal
medals_by_country = Athens_with_medals.groupby('Team')['Medal'].count().reset_index()

# Sort by the number of medals in descending order
top_countries = medals_by_country.sort_values(by='Medal', ascending=False).head(5)

# Create a bar plot for the top 5 countries with the most medals
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Medal', data=top_countries, palette='Set1')
plt.title('Top 5 Countries with the Most Medals')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.show()
```

```python
# Barplot of number of athletes in each Olympic game
athletes_per_game = Athens_with_medals.groupby('Year')['ID'].nunique().reset_index()

# Create a bar plot for the number of athletes in each Olympic game
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='ID', data=athletes_per_game, palette='viridis')
plt.title('Number of Athletes in Each Olympic Game')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# The number of althletes participating in the olympic games saw an upward and downward movement after 1992. The reason confirms the data history.
# The Olympic games were staggered on a 4 year cycle and split into winter and summer games. More Participants showed up for the Summer games more than the Winter Games.
```

```python
# Boxplot for Age distribution of male/female in Olympic games
# Creating a box plot for age distribution of male and female athletes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sex', y='Age', data=Athens_with_medals, palette='pastel')
plt.title('Age Distribution of Male and Female in Olympic Games')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
# There are outliers in both the male and female age groups. The predominat ages falling between 15 and 40. 
```

```python
# Boxplot for variation of age for female over time
female_athletes = Athens_with_medals[Athens_with_medals['Sex'] == 'F']

# Create a box plot for the variation of age for female athletes over time
plt.figure(figsize=(14, 7))
sns.boxplot(x='Year', y='Age', data=female_athletes, palette='viridis')
plt.title('Variation of Age for Female Athletes Over Time')
plt.xlabel('Year')
plt.ylabel('Age')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Scatterplot for Height and Weight ratio of athletes
# Creating a new column for height-to-weight ratio
Athens_with_medals.loc[:, 'Height_Weight_Ratio'] = Athens_with_medals['Height'] / Athens_with_medals['Weight']

# Create a scatter plot for height and weight ratio of athletes
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Height', y='Weight', hue='Height_Weight_Ratio', data=Athens_with_medals, palette='viridis', size='Height_Weight_Ratio', sizes=(20, 200))
plt.title('Scatter Plot for Height and Weight Ratio of Athletes')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(title='Height-to-Weight Ratio')
plt.show()
```

```python
# The above plot demonstrates the varied physical characteristics of athletes. The height-to-weight ratio provides insights into the diversity of body compositions among athletes.
```

```python
# Heatmap for average age of medal winners in Olympic games.
# Grouping by Olympic game and calculate the average age of medal winners
average_age_by_game = Athens_with_medals.groupby('Year')['Age'].mean().reset_index()

# Create a heatmap for the average age of medal winners
plt.figure(figsize=(12, 6))
sns.heatmap(data=average_age_by_game.pivot_table(index='Year', values= 'Age'), cmap='viridis', annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Age of Medal Winners in Olympic Games')
plt.xlabel('Age')
plt.ylabel('Olympic Year')
plt.show()
```
